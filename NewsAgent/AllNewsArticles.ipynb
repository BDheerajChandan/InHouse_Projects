```python
from newspaper import news_pool, build
import pprint
import time
news_channels = {
    "Indian News Channels": {
        "NDTV": "https://www.ndtv.com/",
        "Times of India": "https://timesofindia.indiatimes.com/",
        "India Today": "https://www.indiatoday.in/",
        "The Indian Express": "https://indianexpress.com/",
        "The Hindu": "https://www.thehindu.com/",
        "News18": "https://www.news18.com/",
        "Hindustan Times": "https://www.hindustantimes.com/",
        "Zee News": "https://zeenews.india.com/",
        "ABP News": "https://www.abplive.com/",
        "Aaj Tak": "https://www.aajtak.in/",
        "Business Standard": "https://www.business-standard.com/",
        "Deccan Chronicle": "https://www.deccanchronicle.com/",
        "DNA India": "https://www.dnaindia.com/",
        "Firstpost": "https://www.firstpost.com/",
        "Outlook India": "https://www.outlookindia.com/",
        "The Telegraph": "https://www.telegraphindia.com/",
        "The Quint": "https://www.thequint.com/",
        "Scroll.in": "https://scroll.in/",
        "The Wire": "https://thewire.in/",
        "The Print": "https://theprint.in/",
        "LiveMint": "https://www.livemint.com/",
        "Free Press Journal": "https://www.freepressjournal.in/",
        "The Pioneer": "https://www.dailypioneer.com/",
        "Oneindia": "https://www.oneindia.com/",
        "Sify News": "https://www.sify.com/news/",
        "The Statesman": "https://www.thestatesman.com/",
        "Mid-Day": "https://www.mid-day.com/",
        "The Tribune": "https://www.tribuneindia.com/",
        "The Economic Times": "https://economictimes.indiatimes.com/",
        "Financial Express": "https://www.financialexpress.com/",
        "Moneycontrol": "https://www.moneycontrol.com/",
        "Business Today": "https://www.businesstoday.in/",
        "NewsX": "https://www.newsx.com/",
        "Republic TV": "https://www.republicworld.com/",
        "Mirror Now": "https://www.timesnownews.com/mirror-now",
        "CNN-News18": "https://www.news18.com/",
        "India TV": "https://www.indiatvnews.com/",
        "WION": "https://www.wionews.com/",
        "TV9 Bharatvarsh": "https://tv9bharatvarsh.com/",
        "News Nation": "https://www.newsnationtv.com/",
        "DD News": "https://ddnews.gov.in/",
        "Lok Sabha TV": "https://loksabhatv.nic.in/",
        "Rajya Sabha TV": "https://rajyasabhatv.nic.in/",
        "News24": "https://news24online.com/",
        "India News": "https://www.indiatvnews.com/",
        "Janata TV": "https://www.janataparty.org/",
        "Tehelka": "https://www.tehelka.com/",
        "The Week": "https://www.theweek.in/",
        "The Sunday Guardian": "https://www.sundayguardianlive.com/",
        "Asianet News": "https://www.asianetnews.com/",
        "Sakshi": "https://www.sakshi.com/"
    },
    "International News Channels": {
        "BBC News": "https://www.bbc.com/",
        "CNN": "https://edition.cnn.com/",
        "Al Jazeera": "https://www.aljazeera.com/",
        "Reuters": "https://www.reuters.com/",
        "The New York Times": "https://www.nytimes.com/",
        "The Guardian": "https://www.theguardian.com/international",
        "The Washington Post": "https://www.washingtonpost.com/",
        "Bloomberg": "https://www.bloomberg.com/",
        "Associated Press": "https://apnews.com/",
        "Fox News": "https://www.foxnews.com/",
        "NBC News": "https://www.nbcnews.com/",
        "CBS News": "https://www.cbsnews.com/",
        "Sky News": "https://news.sky.com/",
        "France 24": "https://www.france24.com/",
        "Deutsche Welle": "https://www.dw.com/",
        "Euronews": "https://www.euronews.com/",
        "Voice of America": "https://www.voanews.com/",
        "RT News": "https://www.rt.com/",
        "China Global Television Network": "https://www.cgtn.com/",
        "NHK World": "https://www3.nhk.or.jp/nhkworld/",
        "CCTV": "http://english.cctv.com/",
        "ABC News (Australia)": "https://www.abc.net.au/",
        "CBC News (Canada)": "https://www.cbc.ca/",
        "The Times": "https://www.thetimes.co.uk/",
        "The Wall Street Journal": "https://www.wsj.com/",
        "Financial Times": "https://www.ft.com/",
        "Politico": "https://www.politico.com/",
        "The Economist": "https://www.economist.com/",
        "Time": "https://time.com/",
        "Newsweek": "https://www.newsweek.com/"
    }
}
def get_category_urls_and_articles(name, url):
    try:
        paper = build(url, memoize_articles=False)
        categories = list(paper.category_urls())

        print(f"\n📰 {name} - {url}")
        print(f"Found {len(categories)} categories:")

        if not categories:
            print("   ⚠️ No category URLs found.")
            return

        for cat_url in categories:
            print(f"\n - 📂 Category: {cat_url}")
            try:
                cat_paper = build(cat_url, memoize_articles=False)
                article_urls = [article.url for article in cat_paper.articles]

                # Build inner categories (nested categories)
                inner_categories = list(cat_paper.category_urls())
                total_inner_articles = 0

                # Process inner categories (1 level deep)
                for inner_cat_url in inner_categories:
                    try:
                        inner_cat_paper = build(inner_cat_url, memoize_articles=False)
                        inner_article_urls = [a.url for a in inner_cat_paper.articles]
                        total_inner_articles += len(inner_article_urls)
                        print(f"     🔁 Inner Category: {inner_cat_url}")
                        print(f"       ➤ Found {len(inner_article_urls)} articles.")
                        for url in inner_article_urls[:]:  # Show a all examples
                            print(f"         🔗 {url}")
                        time.sleep(1)
                    except Exception as e:
                        print(f"     ❌ Failed to build inner category: {e}")

                print(f"     📄 Articles in category: {len(article_urls)}")
                for url in article_urls[:5]:
                    print(f"       🔗 {url}")

                print(f"     📚 Total articles (this + inner): {len(article_urls) + total_inner_articles}")
                time.sleep(1)

            except Exception as e:
                print(f"     ❌ Error in category build: {e}")
    except Exception as e:
        print(f"   ❌ Failed to build main site {name}: {e}")
for region, channels in news_channels.items():
    print(f"\n\n{'='*30}\n{region}\n{'='*30}")
    for name, url in channels.items():
        try:
            get_category_urls_and_articles(name, url)
        except Exception as e:
            print(f"Failed to process {name}: {e}")
```
